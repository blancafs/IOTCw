```python
# Imports
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn
from scipy.signal import find_peaks
```

```python
# First data
df = pd.read_csv('../data_csvs/PDIoT_A_20190918T131131.csv', encoding='utf-8')
print(df.columns)
print(df.shape)
df
```

```python
accels = df.plot(y=['accel_x', 'accel_y', 'accel_z'])
```

```python
gyros = df.plot(y=['gyro_x', 'gyro_y', 'gyro_z'])
```

```python
# Seeing gyro readings individually
gyro_sequences = [df['gyro_x'].values, df['gyro_y'].values, df['gyro_z'].values]
i = 0 
gps = []
for r in gyro_sequences:
    peaks, _ = find_peaks(r, height=0)
    plt.plot(r)
    plt.plot(peaks, r[peaks], "r")
    plt.plot(np.zeros_like(r), "--", color="gray")
    t = 'Gyro_'+str(i)
    plt.title(t)
    i+=1
    plt.show()
    gps.append(plt)
```
